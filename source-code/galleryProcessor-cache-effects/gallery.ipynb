# Gallery of processor cache effects

```python
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline
import numpy as np
```

## Cache hierarchy

```python
hierarchy_timings = np.genfromtxt('cache_hierarchy.txt', delimiter=',', unpack=True, names=True)
```

```python
plt.figure(figsize=(12.0, 8.0))
plt.semilogx(hierarchy_timings['array_size'], hierarchy_timings['time'], basex=2)
plt.xticks(np.logspace(7, 24, num=18, base=2))
plt.xlabel('array size')
plt.ylabel('time');
```

## Cache lines

```python
timings = np.genfromtxt('cache_lines.txt', delimiter=',', names=True, unpack=True)
```

```python
plt.figure(figsize=(12.0, 8.0))
plt.loglog(timings['stride'], timings['time'], basex=2, basey=2)
plt.xlabel(r'K')
plt.ylabel(r'time');
```

## Cache associativity

```python
associativity_timings = np.genfromtxt('cache_associativity.txt', delimiter=',', names=True, unpack=True)
```

```python
len(set(associativity_timings['array_size'])), len(set(associativity_timings['step']))
```

```python
timing_map = associativity_timings['time'].reshape((64, 4999))
```

```python
plt.figure(figsize=(10, 12))
plt.imshow(timing_map[:,: 1200], origin='lower', cmap=cm.coolwarm, aspect=10)
plt.xlabel('step size (8 bytes)')
plt.ylabel('array size (8 byte elements)')
plt.xticks([x for x in range(0, 1100, 128)])
plt.yticks([x for x in range(0, 64, 8)]);
```

```python
plt.figure(figsize=(10, 12))
plt.imshow(timing_map[52:,: 270], origin='lower', cmap=cm.coolwarm, aspect=20)
plt.xlabel('step size')
plt.ylabel('array size');
```

```python
associativity_bytes_timings = np.genfromtxt('cache_associativity_bytes.txt', delimiter=',', names=True, unpack=True)
```

```python
timing_map = associativity_bytes_timings['time'].reshape((64, 5000))
```

```python
np.min(timing_map), np.median(timing_map), np.max(timing_map)
```

```python
plt.figure(figsize=(10, 12))
plt.imshow(timing_map[:,: 1200], origin='lower', cmap=cm.coolwarm, aspect=10, interpolation='none')
plt.xlabel('step size (8 bytes)')
plt.ylabel('array size (8 byte elements)')
plt.xticks([x for x in range(0, 1100, 128)])
plt.yticks([x for x in range(0, 64, 8)]);
```

```python
plt.figure(figsize=(10, 12))
plt.imshow(timing_map[24:56,: 300], origin='lower', cmap=cm.coolwarm, aspect=5, interpolation='none')
plt.xlabel('step size')
plt.ylabel('array size')
plt.xticks([x for x in range(0, 300, 16)]);
```
